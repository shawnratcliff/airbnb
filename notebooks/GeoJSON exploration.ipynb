```python
# Imports
import json
from pprint import pprint
```

```python
# Load geojson data
with open('../res/neighbourhoods.geojson') as f:
    data = json.load(f)
```

```python
# Explore object structure
features = data['features']
print(len(features)) # 272, one for each neighborhood
woodland_hills = features[-1]
pprint(woodland_hills)
wh_coordinates_list = woodland_hills['geometry']['coordinates'][0][0]
print(wh_coordinates_list)
```

```python
# Print all neighborhood names
names=[elem['properties']['neighbourhood'] for elem in features]
pprint(names)
```
